```python
import numpy as np

counter = 0
for ev in range(1000):
    nGenPart = branches["nGenPart"][ev]
    GenPart_pt = branches["GenPart_pt"][ev]
    GenPart_eta = branches["GenPart_eta"][ev]
    GenPart_genPartIdxMother = branches["GenPart_genPartIdxMother"][ev]
    GenPart_pdgId = branches["GenPart_pdgId"][ev]
    GenPart_charge = branches["GenPart_charge"][ev]
    GenPart_status = branches["GenPart_status"][ev]

    # Select D mesons
    mesonsMask = ((abs(GenPart_pdgId) == 411) | (abs(GenPart_pdgId) == 421) | (abs(GenPart_pdgId) == 431)) & (GenPart_pt > 10) & (abs(GenPart_eta) < 2.5)
    mesons = np.arange(nGenPart)[mesonsMask]

    for mes in mesons:
        nStableDaughters = 0

        # Loop over potential daughter particles
        for gp in range(nGenPart):
            if (GenPart_status[gp] == 1) and (abs(GenPart_charge[gp]) == 1) and (GenPart_pt[gp] > 1) and (abs(GenPart_eta[gp]) < 2.5):
                # Check if the current particle is a daughter of the meson
                motherIdx = GenPart_genPartIdxMother[gp]
                while motherIdx != -1:
                    if motherIdx == mes:
                        nStableDaughters += 1
                        break
                    motherIdx = GenPart_genPartIdxMother[motherIdx]

            # If we already found at least 2 stable daughters, no need to check further
            if nStableDaughters >= 2:
                counter += 1
                break

print("Number of D mesons with the specified criteria:", counter)
```
